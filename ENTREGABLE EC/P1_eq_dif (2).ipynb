SOLUCION DE TRABAJO NUMERO 1 DE ECUACIONES DIFERENCIALES SEGUNDA PARTE
Para esta solucion realizó el problema 4 de aplicaciones, sobre circuitos RC y RL
Para estos circuitos se tomaron los siguientes valores:
  R = 500  ohms
  C = 5 mF = 0.005 F
  L = 4 mH = 0.004 H
  A = 14

  Para casos puntuales de simulación se usó:
  RC: w = 1/RC
  RL: w = R/L

Se realizó el proceso de despeje de las ecuaciones diferenciales, para ello se tomó la ecuación diferencial de circuitos RC y RL. Se partio de las siguientes ecuaciones:
1. CIRCUITO RC

$
R \cdot C \cdot \frac{dV_{out}}{dt} + v_{out} = v_{in}
$

2. CIRCUITO RL

$
L \cdot \frac{dV_{out}}{dt} + R \cdot v_{out} = L \cdot \frac{dV_{in}}{dt}
$

```python
# ANALISIS AMPLITUD CIRCUITO RC
# importar librerias

import numpy as np
import sympy as sp
import matplotlib.pyplot as plt

import pandas as pd

a = 14
r = 500
c = 0.0005


w = np.linspace(0, 100, 1000)
amp = a/((1+((r*c*w)**2))**(1/2))

plt.plot(w, amp)
plt.title("AMPLITUD DE RC SEGÚN w")
plt.xlabel("frecuencia angular w")
plt.ylabel("amplitud (v)")
plt.legend()

"""
de las gráficas anteriores se toman las amplitudes también de manera manual:
w = 1.4     amplitud =
"""
```

```python
# ANALISIS AMPLITUD CIRCUITO RL
# importar librerias

import numpy as np
import sympy as sp
import matplotlib.pyplot as plt

import pandas as pd

a = 14
r = 500
l = 0.004


w = np.linspace(0, 1000000, 1000)
amp = (a*w*l)/(r*((1+(((l*w)/(r))**2))**(1/2)))

plt.plot(w, amp)
plt.title("AMPLITUD DE RL SEGÚN w")
plt.xlabel("frecuencia angular w")
plt.ylabel("amplitud (v)")
plt.legend()
```
